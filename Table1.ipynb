```python
import pandas as pd
from pathlib import Path
from IPython.display import display
```

```python
#input_path = Path("ARR_MARCH") / "Output" / "summaries" / "Competence"
input_path = Path("Output") / "summaries" / "Competence"
```

```python
single_culture = ['Japan', 'Armenia', 'Germany', 'Greece', 'Netherlands']
```

```python
columns_needed = [
    'domain', 'country', 'mean_comp_human', 'mean_GPT-OSS_120B', 'mean_LLAMA3_1_70B', 'mean_LLAMA3_70B',
    'mean_PHI3_INSTRUCT', 'mean_QWEN2_5VL_32B', 'mean_QWEN2_5VL_72B',
    'mean_QWEN2_5VL_7B', 'mean_QWEN3_32B', 'mean_QWEN_7B', 'mean_GPT_4O'
]
```

```python
files = list(input_path.glob("*.csv"))
#files[:5]
```

```python
table1_list = []
for file in files: 
    dfread = pd.read_csv(file)
    df1 = dfread[columns_needed]
    dfsingle = df1[df1['country'].isin(single_culture)]
    dfmulti = df1[~df1['country'].isin(single_culture)]
    
    human_comp_single = dfsingle['mean_comp_human'].mean()
    human_comp_multi = dfmulti['mean_comp_human'].mean()
    
    gpt_oss_single = dfsingle['mean_GPT-OSS_120B'].mean()
    gpt_oss_multi = dfmulti['mean_GPT-OSS_120B'].mean()
    
    llama31_single = dfsingle['mean_LLAMA3_1_70B'].mean()
    llama31_multi = dfmulti['mean_LLAMA3_1_70B'].mean()
    
    llama3_single = dfsingle['mean_LLAMA3_70B'].mean()
    llama3_multi = dfmulti['mean_LLAMA3_70B'].mean()
    
    phi3_single = dfsingle['mean_PHI3_INSTRUCT'].mean()
    phi3_multi = dfmulti['mean_PHI3_INSTRUCT'].mean()
    
    qwen25vl32_single = dfsingle['mean_QWEN2_5VL_32B'].mean()
    qwen25vl32_multi = dfmulti['mean_QWEN2_5VL_32B'].mean()
    
    qwen25vl72_single = dfsingle['mean_QWEN2_5VL_72B'].mean()
    qwen25vl72_multi = dfmulti['mean_QWEN2_5VL_72B'].mean()
    
    qwen25vl7_single = dfsingle['mean_QWEN2_5VL_7B'].mean()
    qwen25vl7_multi = dfmulti['mean_QWEN2_5VL_7B'].mean()
    
    qwen332_single = dfsingle['mean_QWEN3_32B'].mean()
    qwen332_multi = dfmulti['mean_QWEN3_32B'].mean()

    qwen7_single = dfsingle['mean_QWEN_7B'].mean()
    qwen7_multi = dfmulti['mean_QWEN_7B'].mean()
    
    gpt_4o_single = dfsingle['mean_GPT_4O'].mean()
    gpt_4o_multi = dfmulti['mean_GPT_4O'].mean()
    
    file_split = file.with_suffix('').name.split('_')
    if len(file_split) == 2:
        domain = file_split[-1]
    else:
        domain = file_split[1] + '_' + file_split[2]
    
    table1_list.append([domain, human_comp_multi, human_comp_single, gpt_oss_multi, gpt_oss_single, llama31_multi, llama31_single,
                       llama3_multi, llama3_single, phi3_multi, phi3_single, qwen25vl32_multi, qwen25vl32_single, qwen25vl72_multi, qwen25vl72_single,
                       qwen25vl7_multi, qwen25vl7_single, qwen332_multi, qwen332_single, qwen7_multi, qwen7_single, gpt_4o_multi, gpt_4o_single])
```

```python
table1_columns = ['domain', 'human_multi', 'human_single', 'gpt_oss_120B_multi',
                  'gpt_oss_120B_single', 'llama31_70B_multi', 'llama31_70B_single',
                  'llama3_70B_multi', 'llama3_70B_single', 'phi3_multi', 'phi3_single',
                  'qwen_25vl_32B_multi', 'qwen_25vl_32B_single', 'qwen_25vl_72B_multi', 'qwen_25vl_72B_single',
                  'qwen_25vl_7B_multi', 'qwen_25vl_7B_single', 'qwen3_32B_multi', 'qwen3_32B_single',
                  'qwen_7B_multi', 'qwen_7B_single', 'gpt_4o_multi', 'gpt_4o_single']
```

```python
dftable1 = pd.DataFrame(table1_list, columns = table1_columns)
```

```python
dftable1
```

```python
num_cols = dftable1.select_dtypes(include="number").columns
dftable1[num_cols] = dftable1[num_cols].round(2)
```

```python
df = dftable1.copy()
```

```python
df["Domain"] = df["domain"].str.replace(r"_(I|II|III)$", "", regex=True)
```

```python
dfagg = (df.groupby("Domain", as_index=False)[num_cols].mean())
```

```python
order = ["EV","EVN","HWB","PCPR","PIPP","POC","POM","POS","POST","RV","SCTOM","SVNS"]
dfagg["Domain"] = pd.Categorical(dfagg["Domain"], categories=order, ordered=True)
dfagg = dfagg.sort_values("Domain").set_index("Domain")
```

```python
multi_cols  = [c for c in dfagg.columns if c.endswith("_multi")]
single_cols = [c for c in dfagg.columns if c.endswith("_single")]
other_cols  = [c for c in dfagg.columns if c not in multi_cols + single_cols]

dfagg = dfagg[multi_cols + single_cols + other_cols]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sc_columns = ['human_single', 'gpt_oss_120B_single', 'llama3_70B_single',
                'qwen_25vl_72B_single', 'gpt_4o_single', 
                'llama31_70B_single', 'phi3_single', 'qwen_25vl_32B_single', 'qwen_25vl_7B_single',
                   'qwen3_32B_single', 'qwen_7B_single']

mc_columns = ['human_multi', 'gpt_oss_120B_multi',
                'llama3_70B_multi', 'qwen_25vl_72B_multi', 'gpt_4o_multi',
                'llama31_70B_multi', 'phi3_multi', 'qwen_25vl_32B_multi',
                   'qwen_25vl_7B_multi', 'qwen3_32B_multi', 'qwen_7B_multi',
                   ]
```

```python
sc_df = dfagg[sc_columns].copy()
mc_df = dfagg[mc_columns].copy()


sc_df.rename(columns=lambda x: x.replace("_single", ""), inplace=True)
mc_df.rename(columns=lambda x: x.replace("_multi", ""), inplace=True)

rename_dict = {
    'human': 'Human',
    'gpt_oss_120B': 'GPT-OSS-120B',
    'llama3_70B': 'Llama3-70B',
    'llama31_70B': 'Llama3.1-70B',
    'phi3': 'Phi3-Instruct',
    'qwen_25vl_32B': 'Qwen2.5VL-32B',
    'qwen_25vl_72B': 'Qwen2.5VL-72B',
    'qwen_25vl_7B': 'Qwen2.5VL-7B',
    'qwen3_32B': 'Qwen3-32B',
    'qwen_7B': 'Qwen-7B',
    'gpt_4o': 'GPT-4o'
}
sc_df.rename(columns=rename_dict, inplace=True)
mc_df.rename(columns=rename_dict, inplace=True)
```

```python

plt.figure(figsize=(6, 6))
ax = sns.heatmap(sc_df, annot=True, vmin=0, vmax=1, cbar=False)
ax.xaxis.tick_top()  # Move x-axis labels to top
ax.set_xticklabels(ax.get_xticklabels(), rotation=270, ha='right')
plt.savefig("Output/plots/heatmap_sc.png", bbox_inches='tight', dpi=300)
plt.show()


```
